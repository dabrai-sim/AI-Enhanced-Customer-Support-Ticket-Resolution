```python
# TF-IDF
# term frequency - inverse document frequency
# (count word occ in row / total words in row) * (no of rows word occ in / no of rows)^-1

#       word1                word2         word3 ...
# row1  wordF11 * wordIDF1   tf2 * idf2
# row2  wordF12 * wordIDF1
# row3  wordF13 * wordIDF1
```

```python
# prompt: get tfidf values for 'answer' column

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# Fit and transform the 'answer' column
tfidf_matrix = vectorizer.fit_transform(df['answer'].astype(str))

# Get the feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Create a DataFrame from the TF-IDF matrix
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
```

```python
# prompt: calc pair wise similarity for tfidf_df nrows x nrows, and plot heatmap

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity

# Assuming tfidf_df is already defined from the previous code
# Calculate pairwise cosine similarity
similarity_matrix = cosine_similarity(tfidf_df)

# Convert the similarity matrix to a DataFrame for easier plotting
similarity_df = pd.DataFrame(similarity_matrix)

# Create a heatmap of the similarity matrix
plt.figure(figsize=(10, 8))
sns.heatmap(similarity_df, cmap="viridis")
plt.title("Pairwise Similarity Heatmap")
plt.xlabel("Document Index")
plt.ylabel("Document Index")
plt.show()
```

```python
# prompt: use pca to reduce dimansions to 3 principal components

from sklearn.decomposition import PCA

# Assuming 'tfidf_df' is your DataFrame with TF-IDF values
# Replace 'tfidf_df' with the actual name of your DataFrame if it's different.

# Create a PCA object with 3 components
pca = PCA(n_components=3)

# Fit the PCA model to your data
pca.fit(tfidf_df)

# Transform your data to the lower-dimensional space
principal_components = pca.transform(tfidf_df)

# Create a new DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])

pca_df.head()
```

```python
# prompt: do kmeans clustering and plot graph for elbow method

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assuming 'tfidf_df' is your DataFrame with TF-IDF values
# Replace 'tfidf_df' with the actual name of your DataFrame if it's different.

# Calculate inertia for different values of k
inertias = []
for k in range(1, 15):  # Test k values from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=0) # Initialize KMeans
    kmeans.fit(tfidf_df)  # Fit the model
    inertias.append(kmeans.inertia_) # Calculate and append inertia


# Plot the elbow method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 15), inertias, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
# prompt: i want to know the most common tag_1,tag_2,tag_3,tag_4,tag_5  and queue for every label and 10 most commonly occuring bigrams in each answer and print any 10 full answers for every label

import nltk
from nltk import ngrams
from collections import Counter

nltk.download('punkt')

def analyze_clusters(df, pca_df, similarity_matrix):
    cluster_analysis = {}
    for cluster in pca_df['cluster'].unique():
        cluster_data = df[df.index.isin(pca_df[pca_df['cluster'] == cluster].index)]
        cluster_analysis[cluster] = {}

        # Most common tags
        tags = ['tag_1', 'tag_2', 'tag_3', 'tag_4', 'tag_5']
        for tag in tags:
            cluster_analysis[cluster][tag] = cluster_data[tag].mode().tolist()

        # Most common queue
        cluster_analysis[cluster]['queue'] = cluster_data['queue'].mode().tolist()

        # 10 most common bigrams in answers
        bigrams = []
        for answer in cluster_data['answer']:
            tokens = nltk.word_tokenize(str(answer).lower())
            bigrams.extend(list(ngrams(tokens, 2)))
        bigram_counts = Counter(bigrams)
        cluster_analysis[cluster]['top_10_bigrams'] = bigram_counts.most_common(10)

        # 10 full answers
        cluster_analysis[cluster]['top_10_answers'] = cluster_data['answer'].sample(n=10, random_state=42).tolist()  # Use random_state for reproducibility

    return cluster_analysis

cluster_analysis = analyze_clusters(df, pca_df, similarity_matrix)
cluster_analysis
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def classify_and_respond(query, model, selected_clusters, templates, template_variables, pca_df, kmeans):
    # Step 1: Embed the query
    query_embedding = model.encode([query])

    # Step 2: Calculate cosine similarity with cluster centroids
    similarities = cosine_similarity(query_embedding, kmeans.cluster_centers_)

    # Step 3: Identify the best matching cluster
    best_cluster = similarities.argmax()

    # Step 4: Check if the cluster is automated
    if best_cluster in selected_clusters:
        # Inject variables into the template
        response = templates[best_cluster].format(**template_variables[best_cluster])
        return response
    else:
        return "Sorry, we couldn't automate a response for your query. Please contact support."
```
