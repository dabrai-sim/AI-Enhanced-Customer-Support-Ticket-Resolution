```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("/content/all_tickets.csv")

# Set up the plotting style
sns.set(style="whitegrid")
```

```python
data.shape
```

```python
data.head(20)
```

```python
# Visualization 1: Distribution of Tickets by Category
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='category', palette='viridis')
plt.title('Ticket Counts by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualization 2: Distribution of Tickets by Urgency
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='urgency', palette='crest')
plt.title('Ticket Counts by Urgency')
plt.xlabel('Urgency Level')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Distribution of Tickets by Impact
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='impact', palette='magma')
plt.title('Ticket Counts by Impact')
plt.xlabel('Impact Level')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Visualization 4: Heatmap of Urgency vs Impact
urgency_impact = pd.crosstab(data['urgency'], data['impact'])
plt.figure(figsize=(8, 6))
sns.heatmap(urgency_impact, annot=True, fmt="d", cmap="coolwarm")
plt.title('Heatmap of Urgency vs Impact')
plt.xlabel('Impact Level')
plt.ylabel('Urgency Level')
plt.tight_layout()
plt.show()
```

```python
# Visualization 5: Ticket Distribution by Business Service (Top 10)
top_services = data['business_service'].value_counts().head(10)
top_services.plot(kind='bar', figsize=(10, 6), color='skyblue', edgecolor='black')
plt.title('Top 10 Business Services by Ticket Count')
plt.xlabel('Business Service ID')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualization 7: Sub-category Analysis (Sub_category1)
sub_category_counts = data['sub_category1'].value_counts().head(10)
sub_category_counts.plot(kind='barh', figsize=(10, 6), color='lightcoral', edgecolor='black')
plt.title('Top 10 Sub-categories (Level 1)')
plt.xlabel('Count')
plt.ylabel('Sub-category ID')
plt.tight_layout()
plt.show()
```

```python
# Visualization 8: Correlation Heatmap for Numeric Columns
numeric_cols = ['urgency', 'impact', 'category']
corr_matrix = data[numeric_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="Blues", fmt=".2f")
plt.title('Correlation Heatmap for Numeric Features')
plt.tight_layout()
plt.show()
```

4. **Sub-category Trends**:

6. **Correlations**:
   - The correlation heatmap identifies relationships between numeric features like urgency and impact, which might reveal systemic inefficiencies or dependencies.

```python
import string
import re
import nltk

from tqdm import trange
from nltk import tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.probability import FreqDist
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('stopwords')
```

```python
def clean(body):

    body = body.lower()
    body = re.sub('[^a-z A-Z 0-9-]+', '', body)
    body = " ".join([word for word in body.split() if word not in stopwords.words('english')])

    return body
data['body'] = data['body'].apply(clean)
data.head(10)
data['body'][0]
```

```python
def corpus(text):
    text_list = text.split()
    return text_list
data['body_lists'] = data['body'].apply(corpus)
data.head(10)
corpus = []
for i in trange(data.shape[0], ncols=150, nrows=10, colour='green', smoothing=0.8):
    corpus += data['body_lists'][i]
len(corpus)
```

```python
mostCommon = Counter(corpus).most_common(10)
mostCommon
```

```python
words = []
freq = []
for word, count in mostCommon:
    words.append(word)
    freq.append(count)
sns.barplot(x=freq, y=words)
plt.title('Top 10 Most Frequently Occuring Words')
plt.show()
```

```python
cv = CountVectorizer(ngram_range=(2,2))

bigrams = cv.fit_transform(data['body'])

count_values = bigrams.toarray().sum(axis=0)
ngram_freq = pd.DataFrame(sorted([(count_values[i], k) for k, i in cv.vocabulary_.items()], reverse = True))
ngram_freq.columns = ["frequency", "ngram"]

sns.barplot(x=ngram_freq['frequency'][:10], y=ngram_freq['ngram'][:10])
plt.title('Top 10 Most Frequently Occuring Bigrams')
plt.show()
```

```python
cv1 = CountVectorizer(ngram_range=(3,3))
trigrams = cv1.fit_transform(data['body'])
count_values = trigrams.toarray().sum(axis=0)
ngram_freq = pd.DataFrame(sorted([(count_values[i], k) for k, i in cv1.vocabulary_.items()], reverse = True))
ngram_freq.columns = ["frequency", "ngram"]
sns.barplot(x=ngram_freq['frequency'][:10], y=ngram_freq['ngram'][:10])
plt.title('Top 10 Most Frequently Occuring Trigrams')
plt.show()
```
